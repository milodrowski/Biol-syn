Model grypy - ćwiczenie 1

```python
N = 10**4
S = 10**4-1
I = 1
R = 0
R0 = 1.5
gamma = 0.25
beta = R0 * (gamma/N)

def TrzyRownania(x,t):
        [S,I,R] = x
        
        dx = np.empty(3)
        dx[0] = - beta * S * I
        dx[1] = beta * S * I - gamma * I
        dx[2] = gamma * I
        return dx


t0   = 0
tmax = 250
dt   = 0.1
time = np.arange(t0,tmax,dt)
sol  = odeint(TrzyRownania,[S,I,R],time)
df   = pd.DataFrame(sol,columns=['S','I','R'])
df.plot()
```

Model SEIR

```python
#Susceptible–Podatny
#Exposed–Zakażony-niezakażający
#Infected/Infectious–Chory/Zakażający
#Recovered/Removed–Ozdrowieniec/Usunięty

N = 10**2
S = 10**2-1
E = 1
I = 0
R = 0
R0 = 1.5
gamma = 1/1.6
beta = 1.9*10**(-2)
sigma = 1/0.8
omega = 1/30

def CzteryRownania(x,t):
        [S,E,I,R] = x
        
        dx = np.empty(4)
        dx[0] = - beta * S * I + omega * R
        dx[1] = beta * S * I - sigma * E
        dx[2] = sigma * E - gamma * I
        dx[3] = gamma * I - omega * R
        return dx


t0   = 0
tmax = 100
dt   = 0.1
time = np.arange(t0,tmax,dt)
sol  = odeint(CzteryRownania,[S,E,I,R],time)
df   = pd.DataFrame(sol,columns=['S','E','I','R'])
df.plot()

Se = gamma/beta
Ee = gamma*omega*(beta*N-gamma)/beta*(sigma*omega+gamma*(sigma+omega))
Ie = sigma*omega*(beta*N-gamma)/beta*(sigma*omega+gamma*(sigma+omega))
Re = sigma*omega*(beta*N-gamma)/beta*(sigma*omega+gamma*(sigma+omega))

print(Se, Ee, Ie, Re)
```
